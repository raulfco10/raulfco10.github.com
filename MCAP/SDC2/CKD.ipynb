```python
import pandas as pd

ckddata = pd.read_csv('kidney_disease.csv')

ckddata.head()
```

```python
def removeString(x):
    if(x=='?'):
        return 0
    else:
        return x

ckddata['classification'] = ckddata['classification'].apply(lambda x:x.strip())
ckddata['pcv'] = ckddata['pcv'].fillna(0)
ckddata['pcv'] = ckddata['pcv'].apply(lambda x: removeString(x))
groupBy = ckddata.groupby('classification')
```
