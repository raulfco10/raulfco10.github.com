```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importar el dataset
dataset = pd.read_csv('BTC_historical.csv')
```

```python
X=dataset.iloc[:,6:7].values #Todas las columnas excepto la ultima
y=dataset.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0) #Por la cantidad tan pequena de la muestra se usara un 33% de texting
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regression.predict(X_train), color = "blue")
plt.title("Precio vs Volumen (Conjunto de entrenamiento)")
plt.xlabel("Volumen")
plt.ylabel("Precio (USD)")
plt.show()
```

```python
plt.scatter(X_test, y_test, color = "red")
plt.plot(X_train, regression.predict(X_train), color = "blue")
plt.title("Precio vs Volumen (Conjunto de testing)")
plt.xlabel("Volumen")
plt.ylabel("Precio (USD)")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('Coeficiente de regresion: ', regression.coef_)
```

```python
print('Termino independient, osea el valor que corta al eje Y, X=0', regression.intercept_)
```

```python
print('Error cuadratico medio: ', mean_squared_error(y_train, y_pred))
```
